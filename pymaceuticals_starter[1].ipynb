## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.
#Capomulin, Ketapril, Ramicane, Naftisol
```

```python
#determine the 4 most promising treatements by comparing the mean final tumor volume of each drug treatment. 
final_time_points = mouse_study.loc[mouse_study["Timepoint"] == 45 ]
final_time_points.groupby(["Drug Regimen"]).mean()
```

```python
#Capomulin Quartiles
ftp_capomulin = mouse_study.loc[mouse_study["Drug Regimen"] == "Capomulin"]
capomulin = ftp_capomulin["Tumor Volume (mm3)"]
quartiles = capomulin.quantile([.25, .5, .75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
#Get the Capomulin IQR
iqr = upperq -lowerq
#Identify the Outliers
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
display (capomulin)
f"Upper Outliers: {capomulin.loc[capomulin > upper_bound]}"
#print (iqr)
```
